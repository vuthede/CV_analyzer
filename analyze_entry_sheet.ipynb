```python
import pandas as pd
from pprint import pprint
from googletrans import Translator
from collections import Counter
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
print(df)
```

```python
df_groupby_type = df.groupby(['いつの選考ですか'])
groups = df_groupby_type.groups
keys = groups.keys()
```

```python
print("In {} ({}):\n".format('このESはES選考を通過しましたか',
                             Translator().translate('このESはES選考を通過しましたか').text))
for key in keys:
    group = df_groupby_type.get_group(key)
    es_dicision = list(group['このESはES選考を通過しましたか'])

    letter_counts = Counter(es_dicision)
    value_most, count_most = letter_counts.most_common(1)[0] 
    print("For {}({}):\nValues most: {} --> {}\n".format(key, Translator().translate(key).text,
                                                     value_most, Translator().translate(value_most).text))
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    df.plot(kind='bar')
    plt.title("Histogram of  of {}".format(Translator().translate(key).text))
    plt.xticks([])
```

```python
print("In {} ({}):\n".format('この企業の最終的な選考結果を教えてください',
                           Translator().translate('この企業の最終的な選考結果を教えてください').text))
for key in keys:
    group = df_groupby_type.get_group(key)
    es_dicision = list(group['この企業の最終的な選考結果を教えてください'])

    letter_counts = Counter(es_dicision)
    value_most, count_most = letter_counts.most_common(1)[0] 
    print("For {}({}):\nValues most: {} --> {}\n".format(key, Translator().translate(key).text,
                                                     value_most, Translator().translate(value_most).text))
    
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    df.plot(kind='bar')
    plt.title("Histogram of Selection Result of {}".format(Translator().translate(key).text))
    plt.xticks([])


```
